```python
Q1 = df['result'].quantile(0.25)
Q3 = df['result'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df['result'] >= lower) & (df['result'] <= upper)]
sns.boxplot(df['result'])
```

```python
def assign_age_grps(age):
    if age < 4: 
        return 'Toddler'
    elif age < 12:
        return 'Kid'
    elif age < 18: 
        return 'Teenager'
    elif age < 40: 
        return 'Young'
    else:
        return 'Senior'
    
df['age_grp'] = df['age'].apply(assign_age_grps)
df.head()
```

```python
sns.countplot(x=df['age_grp'], hue=df['Class/ASD'])
plt.show()
```

- young and toddler age grp gave lower chance of having autism. 

```python
def add_feature(data):
  data['sum_score'] = 0
  for col in data.loc[:,'A1_Score':'A10_Score'].columns:
    data['sum_score'] += data[col]
  return data

df = add_feature(df)
```

```python
sns.countplot(x=df['sum_score'], hue=df['Class/ASD'])
plt.show()
```

higher the score higher the risk of autism

```python
# cat_cols = ['gender', 'ethnicity', 'contry_of_res', 'relation']
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df = df.astype(int)
```

```python
X = df.drop('Class/ASD', axis=1)
y = df['Class/ASD']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from xgboost import XGBClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Naive Bayes": BernoulliNB(),
    "XGBoost": XGBClassifier()
}
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(name, ":", acc)
    print(name, "\n", classification_report(y_test, preds))
```

- accuracy highest in random forest and svm
- also, here recall is imp as false negatives are costly in this case. 
- so we further choose random forest and svm to perform hyperparameter tuning and further improving accuracy. 

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100,200,300],
    'max_depth': [None,5,10],
    'min_samples_split': [2,5,10]
}
grid = GridSearchCV(
    RandomForestClassifier(),
    params,
    cv=5,
    scoring='accuracy'
)

grid.fit(X_train, y_train)
print(grid.best_params_)
best_model = grid.best_estimator_
preds = best_model.predict(X_test)
print(accuracy_score(y_test, preds))
print(classification_report(y_test, preds))
```

```python
params_svm = {
    'C': [0.1, 0.2, 0.5, 1.0, 1.5, 2.0],
    'kernel': ['linear', 'rbf', 'sigmoid'], 
    'class_weight': ['balanced', None]
}
grid_svc = GridSearchCV(
    SVC(),
    params_svm,
    cv=5,
    scoring='accuracy'
)

grid_svc.fit(X_train, y_train)
print(grid_svc.best_params_)
best_model_svc = grid_svc.best_estimator_
preds = best_model_svc.predict(X_test)
print(accuracy_score(y_test, preds))
print(classification_report(y_test, preds))
```

both recall(for ASD) and accuracy is better in the case of Random Forest Classifier.
